## Pymaceuticals Activity Summary

#### 1. Among all four treatments, only Capomulin can reduce the volume of tumor during the treatment period. In fact, Infubinol and Ketapril make no significant difference from the placebo.
#### 2. Although Capomulin treatment can effectively reduce the tumor size, all four treatments cannot prevent metastatic spread, with Capomulin having the strongest ameliorating effect though.
#### 3. The survival rate after 45 days are as high as around 85% for Capomulin, and the tendency seems to be static as time goes. However, more investigation should be done to give a final conclusion on this. Overall, based on the analysis, Capomulin is the only treatment worth to be investigated further.

## Analytics

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
df1 = pd.read_csv(mouse_drug_data_to_load)
df2 = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
df = df1.merge(df2, on = "Mouse ID")

# Display the data table for preview
df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

group = df.groupby(["Drug", "Timepoint"], as_index=False)
group_mean = group["Tumor Volume (mm3)"].mean()

# Preview DataFrame
group_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
group_sem = df.groupby(["Drug", "Timepoint"]).sem().drop(columns = ['Mouse ID', 'Metastatic Sites'])
# Convert to DataFrame
group_sem.reset_index(inplace=True)
group_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_mean = group_mean.pivot(index ='Timepoint', columns ='Drug', values = 'Tumor Volume (mm3)')
# Preview that Reformatting worked
reformat_mean
```

```python
# Generate the Plot (with Error Bars)
error_pv = group_sem.loc[(group_sem['Drug'] == 'Placebo'), 'Tumor Volume (mm3)']
error_cv = group_sem.loc[(group_sem['Drug'] == 'Capomulin'), 'Tumor Volume (mm3)']
error_iv = group_sem.loc[(group_sem['Drug'] == 'Infubinol'), 'Tumor Volume (mm3)']
error_kv = group_sem.loc[(group_sem['Drug'] == 'Ketapril'), 'Tumor Volume (mm3)']

plt.errorbar(reformat_mean.index, reformat_mean["Placebo"], fmt = '--d', yerr = error_pv, label = 'Placebo')
plt.errorbar(reformat_mean.index, reformat_mean["Capomulin"], fmt = '--D', yerr = error_cv, label = 'Capomulin')
plt.errorbar(reformat_mean.index, reformat_mean["Infubinol"], fmt = '--^', yerr = error_iv, label = 'Infubinol')
plt.errorbar(reformat_mean.index, reformat_mean["Ketapril"], fmt = '--s', yerr = error_kv, label = 'Ketapril')
plt.legend(loc="upper left")
plt.grid(axis = 'y')
plt.title("Tumor Response to Treatment")
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
#Save Figure
plt.savefig("treatment.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
group_mmean = group["Metastatic Sites"].mean()

# Preview DataFrame
group_mmean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
group_msem = df.groupby(["Drug", "Timepoint"]).sem().drop(columns = ['Mouse ID', 'Tumor Volume (mm3)'])
# Convert to DataFrame
group_msem.reset_index(inplace=True)
# Preview DataFrame
group_msem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_mmean = group_mmean.pivot(index ='Timepoint', columns ='Drug', values = 'Metastatic Sites')
# Preview that Reformatting worked
reformat_mmean.head()
```

```python
# Generate the Plot (with Error Bars)
error_pm = group_msem.loc[(group_msem['Drug'] == 'Placebo'), 'Metastatic Sites']
error_cm = group_msem.loc[(group_msem['Drug'] == 'Capomulin'), 'Metastatic Sites']
error_im = group_msem.loc[(group_msem['Drug'] == 'Infubinol'), 'Metastatic Sites']
error_km = group_msem.loc[(group_msem['Drug'] == 'Ketapril'), 'Metastatic Sites']

plt.errorbar(reformat_mmean.index, reformat_mmean["Placebo"], fmt = '--d', yerr = error_pm, label = 'Placebo')
plt.errorbar(reformat_mmean.index, reformat_mmean["Capomulin"], fmt = '--D', yerr = error_cm, label = 'Capomulin')
plt.errorbar(reformat_mmean.index, reformat_mmean["Infubinol"], fmt = '--^', yerr = error_im, label = 'Infubinol')
plt.errorbar(reformat_mmean.index, reformat_mmean["Ketapril"], fmt = '--s', yerr = error_km, label = 'Ketapril')
plt.legend(loc="upper left")
plt.grid(axis = 'y')
plt.title("Metastatic Spread During Treatment")
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Met. Sites')
#Save Figure
plt.savefig("spread.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

group_count = group["Mouse ID"].count()
group_count = group_count.rename(columns = {"Mouse ID": "Mouse Count"})
# Preview DataFrame
group_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_count = group_count.pivot(index ='Timepoint', columns ='Drug', values = 'Mouse Count')
# Preview the Data Frame
reformat_count.head()
```

```python
# Generate the Plot (Accounting for percentages)
pct_count = reformat_count.div(reformat_count.iloc[0, :])*100

plt.plot(pct_count.index, pct_count["Placebo"], '--d', label = 'Placebo')
plt.plot(pct_count.index, pct_count["Capomulin"], '--D', label = 'Capomulin')
plt.plot(pct_count.index, pct_count["Infubinol"], '--^', label = 'Infubinol')
plt.plot(pct_count.index, pct_count["Ketapril"], '--s', label = 'Ketapril')
plt.legend(loc="lower left")
plt.grid()
plt.title("Survival During Treatment")
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
#Save Figure
plt.savefig("survival.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percentage = (group_mean.groupby("Drug").last()['Tumor Volume (mm3)'] - group_mean.groupby("Drug")\
 .first()['Tumor Volume (mm3)'])/group_mean.groupby('Drug').first()['Tumor Volume (mm3)'] * 100
# Display the data to confirm
percentage = pd.DataFrame(percentage).reset_index()
percentage

# Store all Relevant Percent Changes into a Tuple
pct_need = percentage.loc[(percentage['Drug'] == 'Capomulin') | (percentage['Drug'] =='Infubinol')
                           | (percentage['Drug'] =='Ketapril') | (percentage['Drug'] =='Placebo'), :]

# Splice the data between passing and failing drugs
pct_need["Pass"] = pct_need['Tumor Volume (mm3)'] < 0
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar("Drug", "Tumor Volume (mm3)", data=pct_need, color=["g","r","r","r"])

plt.xlim([-1, 4])
plt.grid()
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor Change Over 45 Day Treatment')

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 0.8*height,
                '%s %%\n' % int(height),
        ha='center', va='bottom')
autolabel(rects1)

# Save figure
plt.savefig('change.png')

#Show figure
plt.show()
```
